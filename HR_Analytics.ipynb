```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Dropping duplicates
df_cp = df.drop_duplicates()

# Removing NaN values
df_cp = df.dropna()
```

```python
df_cp
```

```python
features = df_cp.select_dtypes(include=[np.number])
features
```

```python
# Showing a correlation map for all numeric variables
corr_matrix = features.corr()
plt.figure(figsize=(20,10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Map for Numeric Variables')
plt.show()
```

```python
# Create a second df copy
df_data = df.copy()
```

```python
data=df_data
```

```python
# Print columns
print(df_data.columns)
```

```python
## Visualize the categorical variables

# Overtime
sns.countplot(data=df_data, x='OverTime')
plt.title('Overtime')
plt.show()
```

```python
# Marital Status
sns.countplot(data=df_data, x='MaritalStatus')
plt.title('Marital Status')
plt.show()
```

```python
# Job Role
plt.figure(figsize=(10,5))
sns.countplot(data=df_data, y = 'JobRole')
plt.title('Job Role')
plt.show()
```

```python
# Gender
sns.countplot(data=df_data, x='Gender')
plt.title('Gender')
plt.show()
```

```python
# Education Field
plt.figure(figsize=(10,5))
sns.countplot(data=df_data, y = 'EducationField')
plt.title('Education Field')
plt.show()
```

```python
# Department
sns.countplot(data=df_data, x='Department')
plt.show()
```

```python
# Business Travel
sns.countplot(data=df_data, x='BusinessTravel')
plt.title('Business Travel')
plt.show()
```

```python
# Relation between Overtime and age
sns.boxplot(data=df_data, x='OverTime', y='Age')
plt.title('Relationship between Overtime and Age')
plt.show()
```

# Plotting numerical variables

```python
# Total working years
sns.histplot(data=df_data, x='TotalWorkingYears', bins=10, kde=True)
plt.title('Total Working Years')
plt.show()
```

```python
# Education Level
plt.title('Education Level')
sns.histplot(data=df_data, x='Education', bins=10, kde=True)
plt.show()
```

```python
# Number of companies worked
sns.histplot(data=df_data, x='NumCompaniesWorked', bins=10, kde=True)
plt.title('Number of Companies Worked')
plt.show()
```

```python
# Distance from Home
plt.title('Distance From Home')
sns.histplot(data=df_data, x='DistanceFromHome', bins=10, kde=True)
plt.show()
```

# Salary Analysis

```python
# Monthly Income Distribution
sns.histplot(data=df_data, x='MonthlyIncome', bins=10, kde=True)
plt.title('Monthly Income')
plt.show()
```

```python
# Salary by Job Role
plt.figure(figsize=(10,5))
sns.boxplot(data=df_data, x='JobRole', y='MonthlyIncome')
plt.title('Salary by Job Role')
plt.xticks(rotation = 90)
plt.show()
```

# Employee satisfaction and Performance

```python
# Job satisfaction distribution
sns.histplot(data=df_data, x='JobSatisfaction', bins=10, kde=True)
plt.title('Job Satisfaction Distribution')
plt.show()
```

```python
# Performance Rating Distribution
sns.histplot(data=df_data, x='PerformanceRating', bins=10, kde=True)
plt.title('Performance Rating Distribution')
plt.show()
```

```python
# Employee Tenure
sns.histplot(data=df_data, x='YearsAtCompany', bins=10, kde=True)
plt.title('Employee Tenure Distribution')
plt.show()
```

```python
# Age Distribution by Department
plt.figure(figsize=(10,6))
sns.boxplot(data=df_data, x='Department', y='Age')
plt.title('Age Distribution by Department')
plt.show()
```

```python
# Attrition by Job Role
plt.figure(figsize=(10,5))
sns.countplot(data=df_data, x='JobRole', hue='Attrition')
plt.xticks(rotation=90)
plt.title('Attrition by Job Role')
plt.show()
```
